```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
base_dir = '../../data/model_ready/csv/'
```

```python
languages = ['croatian', 'dutch', 'english', 'finnish', 'italian', 'spanish']
```

```python
df_dict = {}
for language in languages:
    path = base_dir + 'processed_'+language+'.csv'
    df_dict[language] = pd.read_csv(path)
```

```python
word_dict = {}
phoneme_dict = {}
for key in df_dict.keys():
    word_dict[key] = df_dict[key]['word'].unique()
    phoneme_dict[key] = df_dict[key][df_dict[key]['label'] == 'stop']['phonemes'].values
```

```python
phoneme_lengths = {}
for key in df_dict.keys():
    phoneme_lengths[key] = [(len(item.split(' '))-1) for item in phoneme_dict[key]]
```

```python
fig, axs = plt.subplots(2, 3, figsize=(18,16))
axs[0, 0].hist(phoneme_lengths['croatian'])
axs[0, 0].set_title("Croatian Phoneme Length")
axs[0, 1].hist(phoneme_lengths['dutch'])
axs[0, 1].set_title("Dutch Phoneme Length")
axs[0, 2].hist(phoneme_lengths['english'])
axs[0, 2].set_title("English Phoneme Length")

axs[1, 0].hist(phoneme_lengths['finnish'])
axs[1, 0].set_title("Finnish Phoneme Length")
axs[1, 1].hist(phoneme_lengths['italian'])
axs[1, 1].set_title("Italian Phoneme Length")
axs[1, 2].hist(phoneme_lengths['spanish'])
axs[1, 2].set_title("Spanish Phoneme Length")

fig.tight_layout()
save_file = '../../data/phoneme_len_hist.png'
fig.savefig(save_file, dpi = fig.dpi, facecolor=(1,1,1,1))
```
